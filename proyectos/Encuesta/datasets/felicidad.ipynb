```python
import pandas as pd
```

```python
data = pd.read_csv("grau-de-satisfacci-de-vi(3)-refine.csv")
data
```

```python
data.info()
```

```python
data['anyo'] = data['DateTime'].str[:4] #Extraer solo los primeros cuatro caracteres
data
```

```python
data = data[['anyo','Puntuacion']]
data
```

```python
data ['Puntuacion'] = data['Puntuacion'].str.replace(',', '.').astype(float) #Sustituye coma por punto y convierte a tipo float
```

```python
data ['Puntuacion'] = data['Puntuacion'].round(2)
data
```

```python
data.to_csv('output/felicidad_pandas.csv')
```
