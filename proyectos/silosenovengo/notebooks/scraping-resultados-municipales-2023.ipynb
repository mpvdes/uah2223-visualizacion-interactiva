# Meterlo todo en una función

La siguiente función nos permite hacer una tabla como la anterior cambiando el correspondiente 'uuid', que es el código identificador de cada municipio en la web.

```python
def parseaMunicipio(uuid):
    url = 'http://elecciones.mir.es/resultadoslocales2019//MUN2019/r/resultados/MUN2019/' + uuid + '.json'
    r = requests.get(url)
    resultados = r.json()
    poblacion = resultados['contexto']['nombre']
    candidaturas = resultados['convocatoriaActual']['candidaturas']
    votos = []
    for candidatura in candidaturas:
        votos.append(candidatura['votos']['total'])
    porcentaje = []
    for candidatura in candidaturas:
        porcentaje.append(candidatura['votos']['porcentaje'])    
    df = pd.DataFrame(candidaturas)
    df['votos'] = votos
    df['porcentaje'] = porcentaje
    df['poblacion'] = poblacion
    return df[['poblacion','siglas','siglasPadre','escaniosObtenidos','votos', 'porcentaje']]
```

```python
parseaMunicipio('d_329B-0SGq6Rx6gJUxFxg') #añadiendo el uuid correspondiente nos devuelve la tabla que queremos de cada municipio
```

# Pillar todos los uuid de cada uno de los municipios

```python
# En inspeccionar>network hay un archivo JSON donde aparece toda la info sobre los municipios de madrid
r = requests.get('http://elecciones.mir.es/resultadoslocales2019//geodata/MUN2019/PROV/madrid.json')
```

```python
r.json()
```

```python
municipios = r.json()['hijos'] #En 'hijos' tenemos toda la información que necesitamos del municipio
```

```python
df = pd.DataFrame(municipios) #Creamos un dataframe con todos los municipios
df
```

```python
uuids = df['uuid']
```

```python
uuids #nos muestra el código identificador de cada municipio
```

```python
parseaMunicipio(uuids[67]) #si aplicamos la función y ponemos el número del municipio, nos sale su info
```

```python
from tqdm.notebook import tqdm #esto sirve para que aparezca una barra de progreso en aquello sobre lo que estamos trabajando
```

```python
#vamos a sacar una lista de todos los municipios con su info electoral a partir de los uuid
resultadosElectorales = [] 
for uuid in tqdm(uuids): #se puede hacer sin tqdm y quitando los paréntesis
    #print(uuid)
    resultadoMunicipio = parseaMunicipio(uuid)
    resultadosElectorales.append(resultadoMunicipio)
```

```python
resultadosElectorales
```

```python
resultadosElectorales[150]
```

```python
resultadosTotales = pd.concat(resultadosElectorales) #concatenamos todos los municipios
resultadosTotales
```
